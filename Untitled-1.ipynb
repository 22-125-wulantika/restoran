```python
# Import library yang dibutuhkan
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# 1. Load dataset
data = pd.read_excel('ds.xlsx')  # Ganti dengan path file dataset Anda

# 2. Label Encoding untuk kolom kategorikal (Preferensi Makanan dan Jenis Suasana)
label_encoder = LabelEncoder()
data['Preferensi Makanan'] = label_encoder.fit_transform(data['Preferensi Makanan'])
data['Jenis Suasana'] = label_encoder.fit_transform(data['Jenis Suasana'])

# 3. Menghapus satuan 'km' di Lokasi Restoran dan mengubah ke tipe numerik (float)
data['Lokasi Restoran'] = data['Lokasi Restoran'].str.replace(' km', '').astype(float)

# 4. Memilih hanya fitur yang akan digunakan
features = data[['Preferensi Makanan', 'Lokasi Restoran', 
                 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko', 'Jenis Suasana']]

# 5. Menghitung Cosine Similarity
similarity_matrix = cosine_similarity(features)

# 6. Mengatur tampilan agar hasil dataset lurus ke samping
pd.set_option('display.width', 1000)  # Lebar tampilan
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.colheader_justify', 'center')  # Meratakan header kolom ke tengah

# 7. Menampilkan dataset yang sudah diubah dan matriks cosine similarity
print("Dataset Setelah Preprocessing:")
print(features.head().to_string(index=False))  # Menghapus index agar lebih rapi

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

# 8. Fungsi rekomendasi restoran berdasarkan indeks restoran
def recommend_restaurants(index, similarity_matrix, data, top_n=3):
    similarity_scores = list(enumerate(similarity_matrix[index]))
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    top_indices = [i[0] for i in sorted_scores[1:top_n+1]]  # Exclude the restaurant itself
    recommended = data.iloc[top_indices]
    return recommended

# 9. Contoh Rekomendasi Restoran
print("\nRekomendasi Restoran untuk Restoran ke-0:")
recommended_restaurants = recommend_restaurants(0, similarity_matrix, data)
print(recommended_restaurants.to_string(index=False))  # Menampilkan hasil rekomendasi rapi tanpa index
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder

# 1. Load dataset
data = pd.read_excel('ds.xlsx')  # Ganti dengan path file dataset

# 2. Preprocessing data
label_encoder = LabelEncoder()

# Label Encoding untuk Preferensi Makanan dan Jenis Suasana
data['Preferensi Makanan'] = label_encoder.fit_transform(data['Preferensi Makanan'])
data['Jenis Suasana'] = label_encoder.fit_transform(data['Jenis Suasana'])

# Menghapus satuan 'km' pada Lokasi Restoran dan konversi ke float
data['Lokasi Restoran'] = data['Lokasi Restoran'].str.replace(' km', '').astype(float)

# Menyimpan hanya kolom yang dibutuhkan
features = data[['Preferensi Makanan', 'Lokasi Restoran', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko', 'Jenis Suasana']]

# 3. Input preferensi pengguna
user_preference = {
    'Preferensi Makanan': 1,        # (contoh: 1 = makanan tertentu)
    'Lokasi Restoran': 2.0,         # Lokasi di 2 km
    'Harga Rata-Rata Makanan di Toko (Rp)': 30000,  # Harga maksimum 30.000
    'Rating Toko': 4.5,             # Rating minimal 4.5
    'Jenis Suasana': 1              # Suasana tertentu (1)
}

# Konversi preferensi pengguna ke DataFrame
user_data = pd.DataFrame([user_preference])

# 4. Menghitung cosine similarity
similarity_matrix = cosine_similarity(features, user_data)

# Menyimpan hasil similarity ke dalam data
data['Similarity'] = similarity_matrix

# 5. Filter restoran dengan rating 4.9 dan harga murah (di bawah nilai tertentu, misalnya 30.000)
filtered_data = data[(data['Rating Toko'] == 4.9) & (data['Harga Rata-Rata Makanan di Toko (Rp)'] <= 30000)]

# 6. Mengurutkan berdasarkan similarity tertinggi
sorted_data = filtered_data.sort_values(by='Similarity', ascending=False)

# 7. Menampilkan restoran terbaik dengan nama
top_recommendations = sorted_data.head(5)  # Ambil 5 restoran teratas

# Tampilkan hasilnya
print("Rekomendasi Restoran dengan Harga Murah dan Rating 4.9:")
print(top_recommendations[['Nama Restoran', 'Similarity', 'Lokasi Restoran', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko', 'Jenis Suasana']])
```

```python
import streamlit as st
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
data = pd.read_excel('ds.xlsx')  # Ganti dengan path file dataset

# 1. Label Encoding untuk Preferensi Makanan dan Jenis Suasana
label_encoder = LabelEncoder()
data['Preferensi Makanan'] = label_encoder.fit_transform(data['Preferensi Makanan'])
data['Jenis Suasana'] = label_encoder.fit_transform(data['Jenis Suasana'])

# 2. Menghapus satuan 'km' pada Lokasi Restoran
data['Lokasi Restoran'] = data['Lokasi Restoran'].str.replace(' km', '').astype(float)

# 3. Kolom Harga Rata-Rata Makanan dan Rating Toko sudah numerik, jadi tidak perlu diubah
data = data[['Nama Restoran', 'Preferensi Makanan', 'Lokasi Restoran', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko', 'Jenis Suasana']]

# Menghitung cosine similarity berdasarkan fitur-fitur
features = data[['Preferensi Makanan', 'Lokasi Restoran', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko', 'Jenis Suasana']]
similarity_matrix = cosine_similarity(features)

# Streamlit UI
st.title("Rekomendasi Restoran")

# Pilih rating dan harga
rating_filter = st.slider('Pilih Rating Restoran', min_value=0.0, max_value=5.0, value=4.9, step=0.1)
price_filter = st.slider('Pilih Harga Maksimal (Rp)', min_value=0, max_value=1000000, value=50000, step=1000)

# Menampilkan restoran yang sesuai dengan rating dan harga
filtered_data = data[(data['Rating Toko'] >= rating_filter) & (data['Harga Rata-Rata Makanan di Toko (Rp)'] <= price_filter)]

# Menampilkan hasil rekomendasi
st.subheader("Restoran yang Disarankan:")
st.write(filtered_data[['Nama Restoran', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko']])

# Pilih restoran untuk melihat restoran terdekat
restoran_terpilih = st.selectbox("Pilih Restoran untuk Melihat Rekomendasi Terdekat", filtered_data['Nama Restoran'])

# Mendapatkan indeks restoran yang dipilih
index_restoran = filtered_data[filtered_data['Nama Restoran'] == restoran_terpilih].index[0]

# Menghitung cosine similarity untuk restoran terpilih
similar_restaurants = list(enumerate(similarity_matrix[index_restoran]))

# Mengurutkan berdasarkan similarity dan memilih 5 teratas
similar_restaurants = sorted(similar_restaurants, key=lambda x: x[1], reverse=True)[1:6]

# Menampilkan restoran yang mirip
st.subheader("Restoran Mirip dengan yang Anda Pilih:")
for i in similar_restaurants:
    restaurant_index = i[0]
    restaurant_name = data.iloc[restaurant_index]['Nama Restoran']
    st.write(f"- {restaurant_name}")

```
